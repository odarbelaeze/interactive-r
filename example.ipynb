## Ejercicio 1

Realice un gráfico de puntos de la esperanza de vida contra el año, coloreado por continente.

```r
library('ggplot2')
ggplot(data=gapminder, aes(x=year, y=lifeExp, colour=continent)) + geom_point()
```

# Ejercicio 2

Calcule la media de esperanza de vida en el continente asiático.

```r
mean(gapminder$lifeExp[gapminder$continent == 'Asia'])
```

Calcule la media de esperanza de vida en el continente americano.
